```python
feature_cols = [l for l in train_df.columns.values if 'g-' in l or 'c-' in l]
control_features = train_df[train_df['cp_type'] == 'ctl_vehicle'][feature_cols].values
n_cf = len(control_features)
```

```python
def vis_control_condition(condition, n=10):
    fig, ax = plt.subplots(n, 2, figsize=(20, 10*n))
    
    for i in range(n):
        idx = np.random.choice(n_cf)
        ax[i,0].plot(control_features[idx])
        
    condition_features = train_df[train_df[condition] == 1][feature_cols].values
    n_cond = len(condition_features)
    
    for i in range(n):
        idx = np.random.choice(n_cond)
        ax[i,1].plot(condition_features[idx])
        
    plt.show()
```

```python
vis_control_condition('acetylcholine_receptor_antagonist')
```

```python
def vis_hist(condition, n=10):
    fig, ax = plt.subplots(n, 2, figsize=(10, 5*n))
    
    for i in range(n):
        ax[i,0].hist(control_features[:, i], 25)
        ax[i,0].set_xlim([-5,5])
        
    condition_features = train_df[train_df[condition] == 1][feature_cols].values
        
    for i in range(n):
        ax[i,1].hist(condition_features[:, i], 25)
        ax[i,1].set_xlim([-5,5])
        
    plt.show()
```

```python
vis_hist('acetylcholine_receptor_antagonist')
```
